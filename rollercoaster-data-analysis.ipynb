# Processing the data

```python
data = dataset[['coaster_name',# 'Length', 'Speed', 
         'Location', 'Status',# 'Opening date','Type', 
       #'Manufacturer', 'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 'Opened', 'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
         'year_introduced',# 'latitude', 'longitude',
         'Type_Main','opening_date_clean',
       # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
         'speed_mph', 'height_value',# 'height_unit', 'height_ft',
         'Inversions_clean', 'Gforce_clean']].copy()
```

```python
data.shape
```

```python
data.head()
```

```python
data.dtypes
```

```python
# Ensure using .loc to avoid SettingWithCopyWarning
data.loc[:, 'opening_date_clean'] = pd.to_datetime(data['opening_date_clean'])
```

```python
data.Inversions_clean.unique()     # to check if it's a binary column
```

```python
data.columns
```

```python
data = data.rename(columns={'coaster_name':'Name',
                            'year_introduced':'Year_introduced',
                            'Type_Main':'Material',
                            'opening_date_clean':'Opening_date',
                            'speed_mph':'Speed',
                            'height_value':'Height',
                            'Inversions_clean':'Inversions',
                            'Gforce_clean':'Gforce'})
```

```python
data.isna().sum()
```

```python
data = data.loc[~data.duplicated(subset=['Name','Location','Opening_date'])].reset_index(drop=True).copy()
```

# Visualizing the data

```python
ax = data['Year_introduced'].value_counts().head(11).plot(kind='bar', title='Top years by coaster introduction count')
ax.set_xlabel('Year')
ax.set_ylabel('Count')
plt.show()
```

```python
ax = data['Speed'].plot(kind='hist', title='Coasters speed distribution')
ax.set_xlabel('Speed(mph)')
ax.set_ylabel('')
plt.show()
```

```python
ax = data['Speed'].plot(kind='kde', title='Coasters speed distribution')
ax.set_xlabel('Speed(mph)')
ax.set_ylabel('')
plt.show()
```

```python
data['Material'].value_counts()
```

```python
plt.scatter(x=data.Speed, y=data.Height)
plt.show()
```

```python
data = data[data['Height'] <= 3000]
```

```python
ax = plt.scatter(x=data.Speed, y=data.Height)
ax = plt.xlabel('Speed (mph)')
ax = plt.ylabel('Height (ft)')
ax = plt.title('Speed vs Height')
plt.show()
```

```python
ax = sns.scatterplot(x='Speed', y='Height', hue='Year_introduced', data=data)
ax.set_title('Speed vs. Height')
plt.show()
```
